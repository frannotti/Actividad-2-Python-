```python
 playlist = [
      {"title": "Bohemian Rhapsody", "duration": "5:55"},
      {"title": "Hotel California", "duration": "6:30"},
      {"title": "Stairway to Heaven", "duration": "8:02"},
      {"title": "Imagine", "duration": "3:07"},
      {"title": "Smells Like Teen Spirit", "duration": "5:01"},
      {"title": "Billie Jean", "duration": "4:54"},
      {"title": "Hey Jude", "duration": "7:11"},
      {"title": "Like a Rolling Stone", "duration": "6:13"},
  ]

total_seg = 0

#guardamos las canciones mas y menos larga 
mas_larga = playlist[0]
mas_corta =playlist[0]
seg_max= 0
seg_min = 9999

for cancion in playlist:
    separo = cancion["duration"].split(":") # separamos en una lista con el .split la duracion 
    minutos = int(separo[0])
    segundos = int(separo[1])

    total = minutos * 60 + segundos #paso a segundos y sumo el total 
    total_seg += total 
    #saco el max y min 
    if(total > seg_max):
        seg_max = total 
        mas_larga = cancion 
    if(total < seg_min):
        seg_min = total
        mas_corta = cancion 

#convertimos los seg en min
minutos_total = total_seg // 60
segundos_total = total_seg % 60 

print(f"La duracion total: {minutos_total}m {segundos_total}s")
print(f'Canción más larga: "{mas_larga["title"]}" ({mas_larga["duration"]})')
print(f'Canción más corta: "{mas_corta["title"]}" ({mas_corta["duration"]})')

        

```
