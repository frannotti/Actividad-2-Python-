```python
mensaje = input("ingrese un mensaje: ")
desplazamiento = int(input("ingrese un desplazamiento: "))
cifrado = ""

for letra in mensaje: #cifrado 
    if letra.isupper():
        posicion = (ord(letra) - ord('A') + desplazamiento) % 26 
        cifrado += chr(posicion + ord('A'))
    elif letra.islower():
        posicion = (ord(letra) - ord('a') + desplazamiento) % 26
        cifrado += chr(posicion +ord('a'))
    else:
        cifrado += letra 

descifrado = ""
for letra in cifrado:
    if letra.isupper():
        posicion = (ord(letra) - ord('A') - desplazamiento) % 26
        descifrado += chr(posicion + ord('A'))
    elif letra.islower():
        posicion = (ord(letra) - ord('a') - desplazamiento) % 26
        descifrado += chr(posicion + ord('a'))
    else:
        descifrado += letra 

print(f"mensaje cifrado: {cifrado}")
print(f"mensaje desifrado: {descifrado}")

```
