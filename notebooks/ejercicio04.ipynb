```python
email = input("Ingrese un email: ")

if email.count("@") != 1:
    print("El email no  es valido")

else: 
    separo = email.split("@")
    parte_1 = separo[0]
    parte_2 = separo[1]
    if len(parte_1) == 0:
        print("el email no es valido")
    elif "." not in parte_2:
        print("el email no es valido")
    elif parte_1.startswith("@") or parte_1.startswith("."):
        print(" el email no es valido")
    elif parte_2.endswith("@") or parte_2.endswith("."):
        print("el email no es valido")
    elif len(parte_2.split(".")[-1]) < 2:
        print(" el email no es valido")
    else:
        print("el email es valido")
        
```
