```python
import random 
nombres = input("Ingrese los participante (separados por coma): ")
lista = nombres.split(",")

for i in range(len(lista)):
    lista[i] = lista[i].strip().lower()

if len(lista) < 3:
    print("debe haber al menos 3 participantes ")

else:
    disponibles = lista.copy()
    print("sorteo de amigos invisibles")
    for per in lista:
        elegir = random.choice(disponibles)
        while elegir == per:
            elegir = random.choice(disponibles)
        disponibles.remove(elegir)
        print(f" {per} -> {elegir} ")
    

```
